### Scrape data from one page and store in a DataFrame

What are the packs that are available and when where they released

```python

## Get a list of the year and name of all card packs
card_pack_list = [ ] # List to store information
card_list = soup.find(class_='t_body')
for strong in card_list.find_all(class_='t_row'):
    date = strong.find(class_='time').text.strip() #get year
    name_pack = strong.p.text # get name
    card_pack_list.append({'Year of Release':date,'Name of Pack':name_pack}) # combine Year and name to get readable format
card_pack_list_df = pd.DataFrame(card_pack_list) # format
```
